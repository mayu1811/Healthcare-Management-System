```python
import pandas as pd
import numpy as np
from faker import Faker

# Create a Faker instance
fake = Faker()

# Define the number of rows in the dataset
num_rows = 1000  # Reduced for easier display

# Create random data for disease associations (assuming each doctor can be associated with multiple diseases)
diseases = ['Psoriasis'
         'Varicose Veins',
         'Typhoid',
         'Chicken pox',
         'Impetigo',
         'Dengue',
         'Fungal infection',
         'Common Cold',
         'Pneumonia',
         'Dimorphic Hemorrhoids',
         'Arthritis',
         'Acne',
         'Bronchial Asthma',
         'Hypertension',
         'Migraine',
         'Cervical spondylosis',
         'Jaundice',
         'Malaria',
         'urinary tract infection',
         'allergy',
         'gastroesophageal reflux disease',
         'drug reaction',
         'peptic ulcer disease',
         'diabetes']
doctor_diseases = [np.random.choice(diseases) for _ in range(num_rows)]

# Create random data for doctor ratings
doctor_ratings = np.random.uniform(1, 5, num_rows)

# Create random data for doctor arrival and departure times (in 24-hour clock format)
doctor_arrival_times = [f'{np.random.randint(8, 12):02d}:00' for _ in range(num_rows)]  # Random arrival between 8 AM and 12 PM
doctor_departure_times = [f'{np.random.randint(13, 18):02d}:00' for _ in range(num_rows)]  # Random departure between 1 PM and 6 PM

# Create random data for contact numbers
doctor_contact_numbers = [fake.phone_number() for _ in range(num_rows)]

# Create a DataFrame to store the dummy data
dummy_data = pd.DataFrame({
    'Doctor_ID': range(1, num_rows + 1),
    'Associated_Diseases': doctor_diseases,
    'Rating': doctor_ratings,
    'Arrival_Time': doctor_arrival_times,
    'Departure_Time': doctor_departure_times,
    'Contact_Number': doctor_contact_numbers
})


dummy_data.to_csv('dummy_dataset.csv', index=False)
```

```python
#Converting Dataframe to csv file
df=pd.read_csv('/content/dummy_dataset.csv')
```

```python
#-->Extracting Data

# df['Arrival_Time'] = pd.to_datetime(df['Arrival_Time'])
# df['Departure_Time'] = pd.to_datetime(df['Departure_Time'])
x='gastroesophageal reflux disease'
y='08:00'
# Filter the DataFrame to only include rows where 'Associated_Diseases' is 'x' and the doctor is available at time 'y'
filtered_df = df[(df['Associated_Diseases'] == x ) & (df['Arrival_Time'] <= y) & (df['Departure_Time'] >= y)]

# If no doctor is available at time 'y', print a message and exit
if filtered_df.empty:
    print("No doctor is available at the specified time.")
else:
    # Find the row with the maximum rating
    best_doctor = filtered_df[filtered_df['Rating'] == filtered_df['Rating'].max()]

    # Get the 'Doctor_ID' and 'Contact_Number' of the best doctor
    best_doctor_id = best_doctor['Doctor_ID'].values[0]
    best_doctor_contact = best_doctor['Contact_Number'].values[0]

    print(f"The doctor with the highest rating who can treat disease {x} and is available at time {y} is Doctor {best_doctor_id}. You can contact them at {best_doctor_contact}.")
```
